```python
#membaca dataset pangan
df_pangan = pd.read_csv("D:/Kuliah/Semester 3/UAS datmin/ketidak_cukupan_pangan_prov.csv")
df_pangan.head(5)
```

```python
#membaca dataset kepadatan penduduk
df_pend = pd.read_csv("D:/Kuliah/Semester 3/UAS datmin/kepadatan_penduduk.csv")
df_pend.head(5)
```

```python
df_pend.info()
```

```python
#membaca dataset anak dengan akte kelahiran
df_akt = pd.read_csv("D:/Kuliah/Semester 3/UAS datmin/persen_anak_dengan_akte_lahir.csv")
df_akt.head(5)
```

```python
#membaca dataset proporsi hidup
df_propo = pd.read_csv("D:/Kuliah/Semester 3/UAS datmin/proporsi_hidup_dibawah_50_persen_median_pendapatan.csv")
df_propo.head(5)
```

```python
df_propo.info()
```

```python
#membaca dataset penduduk terampil
df_terampil = pd.read_csv("D:/Kuliah/Semester 3/UAS datmin/persen_penduduk_trampil_tik.csv")
df_terampil.head(5)
```

```python
#memgabungkan dataset pangan dan akte
df = pd.merge(df_pangan, df_akt, how = 'inner')
df.head(1)
```

```python
df.head(20)
```

```python
#memgabungkan dataset sebelumnya dgn kepadatan penduduk
df = pd.merge(df, df_pend, how = 'inner')
df.head(1)
```

```python
df.head(20)
```

```python
#memgabungkan dataset sebelumnya dgn kepadatan penduduk terampil
df = pd.merge(df, df_terampil, how = 'inner')
df.head(1)
```

```python
df.head(20)
```

```python
#memgabungkan dataset sebelumnya dgn proporsi hidup
df = pd.merge(df, df_propo, how = 'inner')
df.head(1)
```
